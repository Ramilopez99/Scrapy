```python
%%writefile /content/scrapyApp/proyecto_imdb/spiders/spider1.py

import scrapy


class ImdbSpider(scrapy.Spider):
    name = 'spider1'
    initial_url = 'https://www.imdb.com/'
    allowed_domains = ['www.imdb.com']
    start_urls = ['https://www.imdb.com/chart/top/?ref_=nv_mv_250']

    def parse(self, response):
        movies = response.css('tbody.lister-list > tr > td.titleColumn')
        print('----------------------------------RESULTADO:', len(movies))
        for movie in movies:
            title = movie.css('a::text').get()
            href = movie.css('a::attr(href)').get()
            link = self.initial_url + href
            print('MOVIE:', title, link)
            yield{
                'title': title,
                'link': link
            }    
```

```python
%%writefile /content/scrapyApp/proyecto_imdb/spiders/spider2.py
import scrapy

class ImdbSpider(scrapy.Spider):
    name = 'spider2'
    allowed_domains = ['www.imdb.com']
    start_urls = ['https://www.imdb.com/search/title/?genres=comedy&explore=title_type,genres&ref_=adv_prv']
    
    def __init__(self, page=1):
        self.page = page

    def parse(self, response):
        initial_url = 'https://www.imdb.com'
        movies = response.css('div.lister-list > div.lister-item > div.lister-item-content')
        for movie in movies:
            place = movie.css('h3 > span.lister-item-index.unbold.text-primary::text').get()
            imdb_item = {
                'place' : place,
                'title' : movie.css('h3 > a::text').get(),
                'rating' : movie.css('p.sort-num_votes-visible > span:nth-child(2)::text').get(),
                'director' : [],
                'stars' : [],
                'user_review' : None,
                'critics_review' : None
            }
            
            url_page = initial_url + movie.css('h3 > a::attr(href)').get()
            yield scrapy.Request(url = url_page, callback = self.parse_detail, cb_kwargs={'imdb_item':imdb_item} )
            

        if self.page < 5:
            self.page += 1
            next_page = response.css('div.desc > a.lister-page-next.next-page::attr(href)').get()
            yield response.follow(next_page, callback = self.parse)

    def parse_detail(self, response, imdb_item):
        imdb_item['director'] = response.css('div.sc-fa02f843-0.fjLeDR > ul > li:nth-child(2) > div > ul > li:nth-child(1) > a::text').get()
        stars = response.css('ul.ipc-inline-list.ipc-inline-list--show-dividers.ipc-inline-list--inline.ipc-metadata-list-item__list-content.baseAlt')

        for star in stars:
            imdb_item['stars'].append(star.css('li > a::text').get())
    
        imdb_item['user_review'] = response.css('a.ipc-link.ipc-link--baseAlt.ipc-link--touch-target.sc-3ff39621-2.kKwCBH.isReview > span > span.score::text') .get()
        imdb_item['critics_review'] = response.css('div.sc-7643a8e3-5.dEAmiF > ul > li:nth-child(2) > a > span > span.score::text') .get()

        yield {
            'place' : imdb_item['place'],
            'title' : imdb_item['title'],
            'rating' : imdb_item['rating'],
            'director' : imdb_item['director'],
            'stars' : imdb_item['stars'],
            'user_review' : imdb_item['user_review'],
            'critics_review' : imdb_item['critics_review']
        }
```
